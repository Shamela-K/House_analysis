cleaning the "Ticket_price_cr" column

```python
df["Ticket_Price_Cr"] = df["Ticket_Price_Cr"].astype(str).str.replace("₹","").str.replace(" Cr","")
```

convert to numeric 

```python
df["Ticket_Price_Cr"] = pd.to_numeric(df["Ticket_Price_Cr"], errors="coerce")
```

```python
df["Ticket_Price_Cr"].dtype
```

```python
df["Amenity_Score"].fillna(df["Amenity_Score"].mean(),inplace=True)
```

drop the unnecessary column

```python
df.drop("Buyer_Comments",axis=1,inplace=True)
```

Handling Invalid Values and filling the missing values

```python
df.loc[df["Unit_Size_Sqft"] == -1, "Unit_Size_Sqft"] = np.nan
```

```python
df["Unit_Size_Sqft"].fillna(df["Unit_Size_Sqft"].median(), inplace=True)
```

```python
df.loc[df["Ticket_Price_Cr"] < 0, "Ticket_Price_Cr"] = np.nan
```

Total records

```python
query = """select count(*) 
            from luxury_housing 
"""
cursor.execute(query)
result = cursor.fetchone()

print(result)
```

Booking Status Count

```python
query = """
SELECT Booking_Status, COUNT(*)
FROM luxury_housing
GROUP BY Booking_Status
"""

cursor.execute(query)

for row in cursor.fetchall():
    print(row)
```

Average Price by Developer

```python
query = """ 
SELECT Developer_Name, AVG(Ticket_Price_Cr)
FROM luxury_housing
GROUP BY Developer_Name
"""
cursor.execute(query)

for row in cursor.fetchall():
    print(row[0], ":", round(row[1],2))
```
